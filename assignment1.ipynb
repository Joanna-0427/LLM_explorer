# Q1 - read_vocabulary：

```python
import string

def read_vocabulary(filename):
  with open(file_name,'r',encoding = 'utf-8') as f:
    lines = f.readlines()
    total = {}
    # Create a translation table to remove punctuation
    translator = str.maketrans('', '', string.punctuation + string.digits)
    for line in lines:
      line = line.strip().lower()
      # Remove punctuation
      line = line.translate(translator)
      line = line.split()
      if line == []:
        continue
      for word in line:
        if word in total:
          total[word] += 1
        else:
          total[word] = 1

    sorted_keys = [k for k, v in sorted(total.items(), key=lambda kv: kv[1], reverse = True)]
    return sorted_keys
```

# Q2 - autocomplete_word:

```python
def process_data(word_list):
  total = {}
  for word in word_list:
    if word in total:
      total[word] += 1
    else:
      total[word] = 1
  return sorted(total.items(),key=lambda kv:kv[1],reverse=True)
```

```python
def autocomplete_word(prefix, model_or_data_structure):
  word_list = []
  for word,count in model_or_data_structure:
    if word.startswith(prefix):
      word_list.append(word)
      if len(word_list) == 10:
        break
  return word_list
```

```python
import streamlit as st
def get_model(file_name):
  return process_data(read_vocabulary(file_name))

def main():
  st.title("Autocomplete App")
  query = st.text_input("Enter a value")
  model = get_model(file_name)
  suggestions = autocomplete_word(query,model_or_data_structure=model)
  st.write(suggestions)

if __name__ == "__main__":
    main()
```
